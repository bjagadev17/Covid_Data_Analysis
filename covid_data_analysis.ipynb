```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_data = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
covid_data
```

```python
covid_data = covid_data.iloc[:-1,1:]
```

```python
covid_data.head(10)
```

```python
covid_data.drop(['new_active', 'new_positive', 'new_death','state_code','new_cured'],inplace=True,axis=1)
```

```python
covid_data
```

```python
covid_data['recovery_rate'] = round((covid_data['cured']/covid_data['positive'])*100,2)
```

```python
covid_data.head()
```

```python
covid_data['fatality_rate'] = round((covid_data['death']/covid_data['positive'])*100,2)
```

```python
covid_data
```

```python
covid_data.isnull().sum()
```

```python
covid_data['recovery_rate'].fillna(0)
```

```python
covid_data['fatality_rate'].fillna(0)
```

```python
max(covid_data['positive'])
```

```python
max(covid_data['fatality_rate'])
```

```python
max(covid_data['recovery_rate'])
```

```python
no_states = np.arange(len(covid_data['state_name']))
```

```python
no_states
```

```python
plt.figure(num=None,figsize=(10,8))
plt.barh(no_states,covid_data['positive'],align='center',color=(1,1,0))
plt.yticks(no_states,covid_data['state_name'])
plt.xlim(1,200000+max(covid_data['positive']))
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
plt.figure(num=None,figsize=(10,8))
plt.barh(no_states,covid_data['active'],align='center',color=(0,1,0))
plt.yticks(no_states,covid_data['state_name'])
plt.xlim(1,10000+max(covid_data['active']))
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
plt.figure(num=None,figsize=(10,8))
plt.barh(no_states,covid_data['death'],align='center',color=(1,0,0))
plt.yticks(no_states,covid_data['state_name'])
plt.xlim(1,1000+max(covid_data['death']))
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
plt.figure(num=None,figsize=(10,8))
plt.barh(no_states,covid_data['recovery_rate'],align='center',color=(1,0,1))
plt.yticks(no_states,covid_data['state_name'])
plt.xlim(1,0.5+max(covid_data['recovery_rate']))
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
plt.figure(num=None,figsize=(10,8))
plt.barh(no_states,covid_data['fatality_rate'],align='center',color=(0,0,1))
plt.yticks(no_states,covid_data['state_name'])
plt.xlim(0,1+max(covid_data['fatality_rate']))
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```
